```python
import pickle
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import frgtools.plotting as frgplt
from matplotlib.patches import Rectangle
```

```python
with open('IRCamImages_4inch.pkl', 'rb') as f:
    d = pickle.load(f)
```

```python
inch_per_px = 4/(62-25) /0.75
```

```python
plt.imshow(d[1][20:100, 45:130], cmap=plt.cm.Spectral_r)
plt.colorbar(label='Temperature (C)')
frgplt.scalebar(.0254*inch_per_px)
```

```python
im = d[2][37:83, 55:120]

fig, ax = plt.subplots(1,2,figsize=(8,4))
plt.sca(ax[1])

plt.imshow(
    im,
    cmap=plt.cm.Spectral_r,
    vmin=0.9*im.max(),
    vmax=im.max()
)
r = Rectangle(
    xy = (22, 15),
    height=3/inch_per_px,
    width=3/inch_per_px,
    color='w',
    fill=None
)

plt.title('Box = 3x3 inches')
plt.gca().add_patch(r)
plt.colorbar(label='Temperature (C)')
frgplt.scalebar(.0254*inch_per_px)


##histogram
hp_sim = im[15:15+21, 22:22+21]
sns.histplot(
    hp_sim[hp_sim>105].ravel()/hp_sim.mean(),
    ax=ax[0]
)
ax[0].set_xlabel('Temperature/Temperature_mean')
```

```python
im = d[2][37:83, 55:120]

fig, ax = plt.subplots(1,2,figsize=(8,4))
plt.sca(ax[1])

plt.imshow(
    im,
    cmap=plt.cm.Spectral_r,
    vmin=0.9*im.max(),
    vmax=im.max()
)

x = 25
y = 20
size = round(2/inch_per_px)
r = Rectangle(
    xy = (x, y),
    height=size,
    width=size,
    color='w',
    fill=None
)

plt.title('Box = 3x3 inches')
plt.gca().add_patch(r)
plt.colorbar(label='Temperature (C)')
frgplt.scalebar(.0254*inch_per_px)


##histogram
hp_sim = im[y:y+size, x:x+size]
sns.histplot(
    hp_sim[hp_sim>105].ravel()/hp_sim.mean(),
    ax=ax[0]
)
ax[0].set_xlabel('Temperature/Temperature_mean')
```
